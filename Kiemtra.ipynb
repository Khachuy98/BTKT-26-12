```
#Câu 2
list1 = ["A","B","C","D","E","F"]
list2 = [1,2,3,4,5,6]
dic = {}
for list1_ in list1:
    for list2_ in list2:
        dic[list1_] = list2_
        list2.remove(list2_)
        break  
 
print (dic)
```

```
#Câu 3
person = {
      "name":"QuocNamm",
      "age":28,
      "Salary":8000,
      "City":"Tuy Hoa"
  }

KeyRemove = ["Salary","City"]

def dele(person,KeyRemove):

  for x in KeyRemove:
      if x in person:
          person.pop(x)

  print(person)

dele(person,KeyRemove)
```
